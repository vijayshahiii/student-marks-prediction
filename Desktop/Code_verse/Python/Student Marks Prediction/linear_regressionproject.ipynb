```python
import pandas as pd
from sklearn.linear_model import LinearRegression  
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  
import numpy as np

# Load dataset
df = pd.read_csv("student_study_data.csv") 

model = LinearRegression()

#Input  &  Output

x = df[['Weekly_Study_Hours']]
y = df['Score']

model.fit(x, y)

Value = float(input("Enter the number of weekly study hours: "))
model_train = model.predict([[Value]])

print(f"Predicted Score for {Value} weekly study hours: {model_train[0]:.2f}")
```
